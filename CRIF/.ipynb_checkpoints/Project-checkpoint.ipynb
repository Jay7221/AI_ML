# Similar Words

```python
import numpy as np
```

```python
your_word = 'scam'

ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10
)
words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
print(distances)
```

```python
doct1 = nlp("Microsoft is fraud")
doct2 = nlp("Microsoft is locate in United States")
```
